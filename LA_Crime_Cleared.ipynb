```python
# Specify columns for analysis
columns_of_interest = ['Date Rptd', 'DATE OCC']
```

```python
# Check for missing values in the specified columns
missing_values = data[columns_of_interest].isnull().sum()
```

```python
# Display missing values information
print("Missing Values in Columns:")
print(missing_values)
```

```python
parse_dates=['Date Rptd']
```

```python
missing_threshold = 0.3
```

```python
columns_to_examine = data.columns[data.isnull().mean() > missing_threshold]
```

```python
if len(columns_to_examine) > 0:
    # Visualize missing values using a heatmap, highlighting columns with more than 30% missing values
    plt.figure(figsize=(12, 8))
    sns.heatmap(data[columns_to_examine].isnull(), cmap='viridis', cbar=False, yticklabels=False)
    plt.title('Columns with More Than 30% Missing Values')
    plt.show()
else:
    print("No columns with more than 30% missing values.")
```

```python
# Examine and decide treatment for each column
for column in columns_to_examine:
    print(f"\nExamining column: {column}")

    # Display unique values and their counts in the column
    print(data[column].value_counts())

    # Assess the nature of missing values and make a decision
    if data[column].dtype == 'object':
        # For categorical columns, replace missing values with the mode
        mode_value = data[column].mode().iloc[0]
        data[column].fillna(mode_value, inplace=True)
        print(f"Updating missing values with mode: {mode_value}")
    else:
        # For numeric columns, replace missing values with the mean
        mean_value = data[column].mean()
        data[column].fillna(mean_value, inplace=True)
        print(f"Updating missing values with mean: {mean_value}")
```

```python
# Display the updated dataset
print("\nUpdated Dataset:")
data.head()
```

```python
# Check if necessary calculated columns need to be added
if 'existing_column1' in data.columns and 'existing_column2' in data.columns:
    # Calculate a new column based on existing columns (example: ratio)
    data['calculated_column'] = data['existing_column1'] / data['existing_column2']
    print("Calculated column 'calculated_column' added.")
```

```python
# Display the updated dataset
print("\nUpdated Dataset:")
print(data.head())
```

```python
if 'amount_column1' in data.columns and 'amount_column2' in data.columns:
    data['amount_ratio'] = data['amount_column1'] / data['amount_column2']
    print("\nCalculated column 'amount_ratio' added.")
```
